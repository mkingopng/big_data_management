```
# start a spark session
spark = SparkSession.builder.master("local").appName("cases").getOrCreate()
```

```
# read the csv file
cases = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/michaelkennethkingston@gmail.com/cases_locations.csv")
```

```
# check
cases.take(5)
```

```
# create a view to query
cases.createOrReplaceTempView("cases")
```

```
# query 
cases_results = spark.sql("SELECT lhd_2010_name AS lhd_names, lhd_2010_code AS lhd_code, notification_date AS date, num FROM cases").show()

# should have no headers? unsure of the sequence of operations for the below:
# calulate maximum daily increase of confirmed cases => max(day_1 - day_0) per lhd
# group by lhd_names or lhd_code
# sort by daily increase desc, date asc, lhd desc
```

```
# save results in the result-sql folder in hdfs
cases_results.write.format("csv").save("dbfs:/FileStore/shared_uploads/michaelkennethkingston@gmail.com/cases_results.csv")
```
